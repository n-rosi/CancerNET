**Keras Modules for defining the CNN**

**Keras** is a high level Deep Learning API (Application Programming Interface), i.e. a software intermediary that allows two applications to talk to each other. It includes:

1. **[Models API](https://keras.io/api/models/)** - to define the Neural networks (NN). The **model class** groups layers into an object with training and inference features.

2. **[Layers API](https://keras.io/api/layers/base_layer/#layer-class)** - building blocks of a NN model (e.g. Dense, Convolution). A layer is a callable object that takes as input one or more tensors and that outputs one or more tensors. They all inherit from the class **layer**.

3. **[Optimizers](https://keras.io/api/optimizers/)** - different methods for doing gradient descent to learn weights of NN (e.g. SGD, Adam).

4. **[Losses](https://keras.io/api/losses/)** - objective functions that the optimizer should minimize for use cases like classification, regression (e.g. categorical_crossentropy, MSE).

Moreover, it provides reasonable defaults for the APIs e.g. learning rates for Optimizers, which would work for the common use cases.

```python
# import the necessary packages
from tensorflow.keras.models import Sequential                
from tensorflow.keras.layers import BatchNormalization                                    
from tensorflow.keras.layers import Conv2D                                      #or SeparableConv2D
from tensorflow.keras.layers import AveragePooling2D          
from tensorflow.keras.layers import MaxPooling2D              
from tensorflow.keras.layers import Activation                
from tensorflow.keras.layers import Flatten                   
from tensorflow.keras.layers import Dropout                   
from tensorflow.keras.layers import Dense                         
from keras.models import load_model               
```

[Sequential](https://keras.io/api/models/sequential/): it is a **class** to create Deep Learning models constituted of a plain stack of layers. 

[BatchNormalization](https://keras.io/api/layers/normalization_layers/batch_normalization/): it is a **class**, it applies a transformation that maintains the mean output close to 0 and the output standard deviation close to 1 and works differently during training and during inference. 

[SeparableConv2D](https://keras.io/api/layers/convolution_layers/separable_convolution2d/): separable convolutions consist of first performing a depthwise spatial convolution (which acts on each input channel separately) followed by a pointwise convolution which mixes the resulting output channels. It is a **class**. 

[AveragePooling2D](https://keras.io/api/layers/pooling_layers/average_pooling2d/): Downsamples the input along its spatial dimensions (height and width) by taking the average value over an input window (of size defined by pool_size) for each channel of the input. The window is shifted by strides along each dimension. It is a **class**. 

[MaxPooling2D](https://keras.io/api/layers/pooling_layers/max_pooling2d/): Downsamples the input along its spatial dimensions (height and width) by taking the maximum value over an input window (of size defined by pool_size) for each channel of the input. The window is shifted by strides along each dimension. It is a **class**. 

[Activation](https://keras.io/api/layers/core_layers/activation/): It is a **class** which applies an activation to the output of the previous layer. However, activation can be passed both through layers and through the argoment of other functions. 

[Flatten](https://keras.io/api/layers/reshaping_layers/flatten/): It a **class** which flattens the input without affecting the batch size. If inputs are shaped (batch,) without a feature axis, then flattening adds an extra channel dimension and output shape is (batch, 1). 

[Dropout](https://keras.io/api/layers/regularization_layers/dropout/): The Dropout layer randomly sets input units to 0 with a frequency of rate at each step during training time, which helps prevent overfitting. Inputs not set to 0 are scaled up by 1/(1 - rate) such that the sum over all inputs is unchanged. It is a **class**. 

[Dense](https://keras.io/api/layers/core_layers/dense/): It is a **class** for general fully connected layers.

**Python Method**

A [method](https://docs.python.org/3/tutorial/classes.html#method-objects) in python is somewhat similar to a function, except it is associated with object/classes. Methods in python are very similar to functions except for two major differences.

1. The method is implicitly used for an object for which it is called.

2. The method is accessible to data that is contained within the class.

**Instance vs Static Methods**

1. **Instance methods** look like normal functions, live inside a class, and require one special parameter, usually called 'self'. When we call a method, we don't pass in a value for this first parameter ourselves. Python takes care of that and passes the caller object as the first argument to the method call. That allows us to access different properties of the object instance within the method.


2. **Static methods** are regular functions that live inside the class in order to indicate a certain level of relation to it. They do not require a special parameter and do not receive the object instance (self) or the class (cls) instance as arguments by default. What they do require instead is the @staticmethod decorator. They are called self-contained methods because they do not have access to the instance or class like the other types of method do.

# Define the CNN model

```python
# step 1: define the model

class CancerNet:
  @staticmethod
  def build(width,height,depth,classes):                                        # Input: (scalar1,scalar2,scalar3,scalar4)
    model=Sequential()                                                          # Creating an object named model from the keras Sequential class

    shape=(height,width,depth)                                                  # initialize the model along with the input shape to be
    channelDim=-1		                                                            # "channels last" and the channels dimension itself
    
    #2DConv => ReLu => Pool => Dropout
    model.add(Conv2D(40, (3,3), padding="valid", input_shape=shape))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=channelDim))
    model.add(AveragePooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))

    #(2DConv => ReLu => Pool) * 2 => Dropout
    model.add(Conv2D(80, (3,3), padding="valid"))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=channelDim))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Conv2D(80, (3,3), padding="valid"))
    model.add(Activation("relu"))
    model.add(BatchNormalization(axis=channelDim))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))

		# FC => RelU => FC => SeLU
    model.add(Flatten())
    model.add(Dense(500,activation="relu"))
    model.add(Dense(250,activation="relu"))                                     #model.add(Activation("selu"))
    model.add(BatchNormalization())
    model.add(Dropout(0.5))

		# softmax classifier
    model.add(Dense(classes,activation="softmax"))                              #model.add(Activation("softmax"))

		# return the constructed network architecture
    return model
```

**Some relevant observations**:

1. _Convolutional layers_:

  A convolutional layer takes as input data with the following shape: \\
  **input** => (batch_size, rows, cols, channels) \\
  The model was expecting a vector with size four, but I passed a size 3. If this occurs the model leaves ambound the batch size, according to the rule **data-item first**: \\
  (batch_size, rows, cols, channels) = (None, 50, 50, 3) \\
  The convolutional layer gives as output: \\
  **Output** => (batch_size, new_rows, new_cols, filters) \\

  The other Input parameters:
  - 40 or 80 filters in the filter banck;
  - (3,3) kernel size;
  - "valid" padding specification;

2. _ReLU_ or _SeLU_:

 Simply takes the output of the convolutional layer and include some non-linearity. It is a scalar functions and return the same dimension of the input.

3. _BatchNormalization_:

  It takes as input the batch to normalize and it returns the same batch but rescaled. The pramaeter "axis=ChannelDim" states that the normlization is carries out on the features axis ([decription here](https://keras.io/api/layers/normalization_layers/batch_normalization/)). 



4. _MaxPooling2D_: 

  It takes the maximum across each 2x2 submatrix. The input parameters are:

  - padding = "valid" for padding (default);
  - stride = "None" (default);
  - (2,2) is the operation dimension;

 **input** => (batch_size, new_rows, new_cols, filters) \\
 **output** => (batch_size, new_rows - pool_size + 1, new_cols- pool_size +1, filters).

5. _Dropout_: 

  This layer randomly sets input units to 0 with a probability of 0.25 or 0.50 (my rate parameters) at each step during training time. which helps prevent overfitting. Inputs not set to 0 are scaled up by 1/(1 - rate) such that the sum over all inputs is unchanged. It works only during training and helps to prevent overfitting.

6. _Flatten_: 

  Flattens the input. Does not affect the batch size. \\
 **input** => (batch_size, new_rows, new_cols, filters) \\
 **output** => (batch_size x new_rows x new_cols x filters,).

7. _Dense_:

   Dense implements the operation "output = activation(dot(input, kernel) + bias)" where:
   - _activation_ is the element-wise activation function passed as the activation argument;
   - _kernel_ is a weights matrix created by the layer;
   - _bias_ is a bias vector created by the layer (only applicable if use_bias is True).

   **input** => ((batch_size x new_rows x new_cols x filters,) \\
   **outpu** => (dense_size, )
